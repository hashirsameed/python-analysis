```python
df["TotalCharges"] = df["TotalCharges"].replace(" ","0")
df["TotalCharges"] = df["TotalCharges"].astype("float")
```

```python
def conv(value):
    if value == 1:
        return "YES"
    else :
        return "NO"
df["SeniorCitizen"] = df["SeniorCitizen"].apply(conv)        
```

```python
df.head()
```

```python
plt.figure(figsize = (3,4))
ax = sns.countplot( x = 'Churn' , data = df)
ax.bar_label(ax.containers[0])
plt.title("Count Of Customer By Churn", fontsize = 10)
plt.show()
```

```python
plt.figure(figsize = (3,3))
gb = df.groupby("Churn").agg({'Churn':"count"})
plt.pie(gb["Churn"], labels=gb.index, autopct="%1.2f%%")
plt.title("Percentage Of Churned Customers" , fontsize = 10)
plt.show()
```

```python
plt.title("Churn By Gender", fontsize = 10)
ax1 = sns.countplot(x = 'gender' , data = df , hue = "Churn")
plt.figure(figsize = (4,4))
ax1.bar_label(ax1.containers[0])
plt.show()
```

```python
ct = df.groupby(['SeniorCitizen', 'Churn']).size().unstack()
ct_norm = ct.div(ct.sum(axis=1), axis=0) * 100  # Convert counts to percentages

# Get Jupyter Default Colors (Orange & Blue)
colors = sns.color_palette("tab10")[:2]  # First two default colors

# Plot Stacked Bar Chart
ax = ct_norm.plot(kind='bar', stacked=True, figsize=(6,4), color=colors)

# Add Labels as Percentages
for bars in ax.containers:
    ax.bar_label(bars, fmt='%.1f%%', label_type='center', fontsize=10, color='white')

# Titles & Labels
plt.title("Churn By SeniorCitizen", fontsize=12)
plt.xlabel("Senior Citizen (0 = No, 1 = Yes)")
plt.ylabel("Percentage (%)")
plt.legend(title="Churn", loc="upper right")
plt.show()
```
